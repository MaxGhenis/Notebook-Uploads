# Functions
A "function" is a block of reusable code used to perform a single action

Each function takes a defined set of inputs and arguments

They can be programmed to return a set output or nothing at all

# Functions

````python
def factorial(x):
    if x == 1:
        return 1
    else:
        return x * factorial(x - 1)
````

````python
x = 6
factorial(x)
````

## Loops
Two types of loops in Python: `for` and `while`

`for` loops run until they hit a specified limit

`while` loops run until a specified condition is no longer met

### `for` Loops
```python
for i in range(start, end):
    <code block>
```

````python
x = 1
for i in range(1, 10):
    x *= i
x
````

Note that the range in the loop is not inclusive. It will only run from 1 to 99

### `for` Loops
You can also iterate over items in a list or list like object using a for loop
```python
for item in list_like:
    <code block>
```

````python
starting_list = [1, 2, 3, 4, 5]
ending_list = []
for item in starting_list:
    ending_list.append(item * 2)
ending_list
````

### `while` Loops
```python
while condition:
    <code block>
```

````python
status = True
x = 1
while status:
    x *= 2
    if x >= 32:
        status = False
    print (x)
x
````

### `while` Loop Examples

`while` loops can also be run as long as your condition is false

````python
status = False
x = 1
while not status:
    x += 10
    if x > 100:
        status = True
    print (x)
x
````

## Classes
Classes are used to create new data types

You can define the methods and attributes associated with your class to define how it behaves

````python
class array():
    def __init__(self, elements=[]):
        self.elements = elements
        self.index = -1
        self.max = len(self.elements) - 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.index == self.max:
            raise StopIteration
        self.index += 1
        return self.elements[self.index]

    def __add__(self, other):
        return array([i + j for i, j in zip(self, other)])

    def __mul__(self, other):
        return array([i * j for i, j in zip(self, other)])

    def __repr__(self):
        return 'array({})'.format([element for element in self.elements])
````

````python
a = array([1, 2, 3])
b = array([4, 5, 6])
````

````python
a + b
````

## Bokeh

A Python interactive visualization library that can be used for anything from plotting time series data to interactive Google Maps

It was created with web browsers in mind

````python
from bokeh.models import (GMapPlot, GMapOptions, Circle, Range1d,
                          ColumnDataSource, PanTool, WheelZoomTool,
                          HoverTool, LassoSelectTool, ResetTool)
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
from bokeh.sampledata.stocks import AAPL, GOOG, IBM, MSFT
output_notebook()
````

### Using Bokeh to Create a Google Map

````python
map_options = GMapOptions(lat=34.858, lng=37.893, map_type='terrain', zoom=7)
plot = GMapPlot(x_range=Range1d(), y_range=Range1d(), map_options=map_options, api_key=keys['syria_refugee_camp']['key'])
plot.title.text = 'Refugee Camp Locations'
#plot.api_key = keys['syria_refugee_camp']['key']
source = ColumnDataSource(camp_sites)
circle = Circle(x='Longitude', y='Latitude', size=15, fill_color='blue', fill_alpha=0.5)
plot.add_glyph(source, circle)
hover = HoverTool(tooltips=[('Name', '@Name'), ('Desc.', '@Descriptio')])
plot.add_tools(PanTool(), WheelZoomTool(), ResetTool(), LassoSelectTool(), hover)
show(plot)
````

````python
def datetime(x):
    return np.array(x, dtype=np.datetime64)
````

````python
AAPL['date'] = datetime(AAPL['date'])
GOOG['date'] = datetime(GOOG['date'])
MSFT['date'] = datetime(MSFT['date'])
aapl_cds = ColumnDataSource(AAPL)
goog_cds = ColumnDataSource(GOOG)
msft_cds = ColumnDataSource(MSFT)
````

## Using Bokeh for Time Series Plots

````python
f = figure(title='Stock Closing Prices', x_axis_type='datetime', width=800, height=500)
f.xaxis.axis_label = 'Date'
f.yaxis.axis_label = 'Price'
f.line('date', 'adj_close', legend='APPL', source=aapl_cds)
f.line('date', 'adj_close', legend='GOOG', color='red', source=goog_cds)
f.line('date', 'adj_close', legend='MSFT', color='green', source=msft_cds)
f.legend.location='top_left'
hover = HoverTool(tooltips=[('Date', '@date{%F}'), ('Price', '@adj_close')],
                  formatters={'date': 'datetime'})
f.add_tools(hover)
show(f)
````
