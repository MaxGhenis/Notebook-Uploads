```python
from taxcalc import *
import pandas as pd
import numpy as np
from bokeh.io import output_notebook
from bokeh.charts import Scatter
from bokeh.plotting import show
output_notebook()
```

```python
recs = Records()
pol = Policy()
calc = Calculator(records=recs, policy=pol)
calc.advance_to_year(2014)
calc.calc_all()
```

```python
# Read in itemized deductions data
item_ded = pd.read_csv('deductions.csv')
```

```python
def distribution(item, weight, agi):
    agi_1 = (((item[agi < 5000] * weight[agi < 5000]).sum() / 1000))
    agi_3 = ((item[(agi > 5000) & (agi < 10000)] * weight[(agi > 5000) & (agi < 10000)]).sum() / 1000)
    agi_4 = ((item[(agi > 10000) & (agi < 15000)] * weight[(agi > 10000) & (agi < 15000)]).sum() / 1000)
    agi_5 = ((item[(agi > 15000) & (agi < 20000)] * weight[(agi > 15000) & (agi < 20000)]).sum() / 1000)
    agi_6 = ((item[(agi > 20000) & (agi < 25000)] * weight[(agi > 20000) & (agi < 25000)]).sum() / 1000)
    agi_7 = ((item[(agi > 25000) & (agi < 30000)] * weight[(agi > 25000) & (agi < 30000)]).sum() / 1000)
    agi_8 = ((item[(agi > 30000) & (agi < 40000)] * weight[(agi > 30000) & (agi < 40000)]).sum() / 1000)
    agi_9 = ((item[(agi > 40000) & (agi < 50000)] * weight[(agi > 40000) & (agi < 50000)]).sum() / 1000)
    agi_10 = ((item[(agi > 50000) & (agi < 75000)] * weight[(agi > 50000) & (agi < 75000)]).sum() / 1000)
    agi_11 = ((item[(agi > 75000) & (agi < 100000)] * weight[(agi > 75000) & (agi < 100000)]).sum() / 1000)
    agi_12 = ((item[(agi > 100000) & (agi < 200000)] * weight[(agi > 100000) & (agi < 200000)]).sum() / 1000)
    agi_13 = ((item[(agi > 200000) & (agi < 500000)] * weight[(agi > 200000) & (agi < 500000)]).sum() / 1000)
    agi_14 = ((item[(agi > 500000) & (agi < 1000000)] * weight[(agi > 500000) & (agi < 1000000)]).sum() / 1000)
    agi_15 = ((item[(agi > 1000000) & (agi < 1500000)] * weight[(agi > 1000000) & (agi < 1500000)]).sum() / 1000)
    agi_16 = ((item[(agi > 1500000) & (agi < 2000000)] * weight[(agi > 1500000) & ( agi < 2000000)]).sum() / 1000)
    agi_17 = ((item[(agi > 2000000) & (agi < 5000000)] * weight[(agi > 2000000) & ( agi < 5000000)]).sum() / 1000)
    agi_18 = ((item[(agi > 5000000) & (agi < 10000000)] * weight[(agi > 5000000) & ( agi < 10000000)]).sum() / 1000)
    agi_19 = ((item[agi > 10000000] * weight[agi > 10000000]).sum() / 1000)
    df = [agi_1, agi_3, agi_4, agi_5, agi_6, agi_7, agi_8, agi_9, agi_10, agi_11,
                       agi_12, agi_13, agi_14, agi_15, agi_16, agi_17, agi_18, agi_19]
    return df
```

```python
tc = pd.DataFrame()
tc['State'] = distribution(calc.records.e18400, calc.records.s006, calc.records.c00100)
tc['RE'] = distribution(calc.records.e18500, calc.records.s006, calc.records.c00100)
tc['IP'] = distribution(calc.records.e19200, calc.records.s006, calc.records.c00100)
tc['Cash'] = distribution(calc.records.e19800, calc.records.s006, calc.records.c00100)
tc['NonCash'] = distribution(calc.records.e20100, calc.records.s006, calc.records.c00100)
tc['Misc'] = distribution(calc.records.e20400, calc.records.s006, calc.records.c00100)
tc['Total'] = distribution(calc.records.c04470, calc.records.s006, calc.records.c00100)
```

```python
index_list = ['Under $5K',
              '$5K-$10K',
              '$10K-$15K',
              '$15K-$20K',
              '$20K-$25K',
              '$25K-$30K',
              '$30K-$40K',
              '$40K-$50K',
              '$50K-$75K',
              '$75K-$100K',
              '$100K-$200K',
              '$200K-$500K',
              '$500K-$1M',
              '$1M-$1.5M',
              '$1.5M-$2M',
              '$2M-$5M',
              '$5M-$10M',
              '$10M and over']
```

```python
soi_state = pd.DataFrame()
soi_state['Percent of State and Local Tax Deductions'] = item_ded['State'] / item_ded['State'].sum()
soi_state['Measure'] = 'SOI'
soi_state['AGI Bin'] = index_list

tc_state = pd.DataFrame()
tc_state['Percent of State and Local Tax Deductions'] = tc['State'] / tc['State'].sum()
tc_state['Measure'] = 'TC'
tc_state['AGI Bin'] = index_list

data_state = [soi_state, tc_state]
plot_data_state = pd.concat(data_state)

plot_state = Scatter(plot_data_state, x='AGI Bin', y='Percent of State and Local Tax Deductions',
                     color='Measure', title='Percent of Total State and Local Tax Deductions')
show(plot_state)
```
