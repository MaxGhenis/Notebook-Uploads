## Distribution of Income Variables - Percentage in Each AGI Bin

```python
sas_dist = pd.DataFrame()
py_dist = pd.DataFrame()
sas_was = distribution(calc_sas.records.e00200, calc_sas.records.s006, calc_sas.records.c00100)
sas_int = distribution(calc_sas.records.e00300, calc_sas.records.s006, calc_sas.records.c00100)
sas_odiv = distribution(calc_sas.records.e00600, calc_sas.records.s006, calc_sas.records.c00100)
sas_qdiv = distribution(calc_sas.records.e00650, calc_sas.records.s006, calc_sas.records.c00100)
sas_biz = distribution(calc_sas.records.e00900, calc_sas.records.s006, calc_sas.records.c00100)
sas_combined = distribution(calc_sas.records.combined, calc_sas.records.s006, calc_sas.records.c00100)
py_was = distribution(calc_py.records.e00200, calc_py.records.s006, calc_py.records.c00100)
py_int = distribution(calc_py.records.e00300, calc_py.records.s006, calc_py.records.c00100)
py_odiv = distribution(calc_py.records.e00600, calc_py.records.s006, calc_py.records.c00100)
py_qdiv = distribution(calc_py.records.e00650, calc_py.records.s006, calc_py.records.c00100)
py_biz = distribution(calc_py.records.e00900, calc_py.records.s006, calc_py.records.c00100)
py_combined = distribution(calc_py.records.combined, calc_py.records.s006, calc_py.records.c00100)
sas_dist['WAS'] = sas_was[1]
sas_dist['INT'] = sas_int[1]
sas_dist['ODIV'] = sas_odiv[1]
sas_dist['QDIV'] = sas_qdiv[1]
sas_dist['BIZ'] = sas_biz[1]
sas_dist['combined'] = sas_combined[1]
sas_dist['AGI Bin'] = index_list()
sas_dist['label'] = 'SAS'
py_dist['WAS'] = py_was[1]
py_dist['INT'] = py_int[1]
py_dist['ODIV'] = py_odiv[1]
py_dist['QDIV'] = py_qdiv[1]
py_dist['BIZ'] = py_biz[1]
py_dist['combined'] = py_combined[1]
py_dist['AGI Bin'] = index_list()
py_dist['label'] = 'Python'
scatter_data = pd.concat([sas_dist, py_dist])
# Create scatter plot objects for each variable
items_tups = [('WAS', 'WAS'), ('INT', 'Interest Income'), ('ODIV', 'Ordinary Dividends'),
              ('QDIV', 'Qualified Dividends'), ('BIZ', 'Business Income'), ('combined', 'Combined Tax Liability')]
scatter_list = list()  # list for scatter plot objects
for item in items_tups:
    title = 'Percent of Total {} by AGI Bin'.format(item[1])
    scatter = Scatter(scatter_data, x='AGI Bin', y=item[0], color='label', ylabel='Percent',
                      title=title, tooltips=[('PCT', '@{}'.format(item[0]))])
    scatter_list.append(scatter)
```

```python
show(column(scatter_list))
```

## Itemized Deduction Amounts

```python
deductions_sas = {'Medical Expenses': (calc_sas.records.e17500 * calc_sas.records.s006).sum(),
                  'State and Local Taxes':  (calc_sas.records.e18400 * calc_sas.records.s006).sum(),
                  'Real Estate Taxes':  (calc_sas.records.e18500 * calc_sas.records.s006).sum(),
                  'Interest Paid':  (calc_sas.records.e19200 * calc_sas.records.s006).sum(),
                  'Charitable Cash Contributions': (calc_sas.records.e19800 * calc_sas.records.s006).sum(),
                  'Charitable Non-Cash Contributions': (calc_sas.records.e20100 * calc_sas.records.s006).sum(),
                  'Total Misc. Expenses':  (calc_sas.records.e20400* calc_sas.records.s006).sum(),
                  'Net Casualty or Loss': (calc_sas.records.e20500 * calc_sas.records.s006).sum()}
ded_sas_df = pd.DataFrame.from_dict(deductions_sas, 'index')
ded_sas_df.columns = ['Total']
ded_sas_df['source'] = 'SAS'

deductions_py = {'Medical Expenses': (calc_py.records.e17500 * calc_py.records.s006).sum(),
                  'State and Local Taxes':  (calc_py.records.e18400 * calc_py.records.s006).sum(),
                  'Real Estate Taxes':  (calc_py.records.e18500 * calc_py.records.s006).sum(),
                  'Interest Paid':  (calc_py.records.e19200 * calc_py.records.s006).sum(),
                  'Charitable Cash Contributions': (calc_py.records.e19800 * calc_py.records.s006).sum(),
                  'Charitable Non-Cash Contributions': (calc_py.records.e20100 * calc_py.records.s006).sum(),
                  'Total Misc. Expenses':  (calc_py.records.e20400* calc_py.records.s006).sum(),
                  'Net Casualty or Loss': (calc_py.records.e20500 * calc_py.records.s006).sum()}
ded_py_df = pd.DataFrame.from_dict(deductions_py, 'index')
ded_py_df.columns = ['Total']
ded_py_df['source'] = 'Python'

ded_full_df = pd.concat([ded_sas_df, ded_py_df])
ded_full_df['index'] = ded_full_df.index
```

```python
ded_bar = Bar(ded_full_df, 'index', 'Total', group='source', title='Itemized Deduction Totals',
              xlabel='Deduction', ylabel='Total', tooltips=[('Deduction', '@index'), ('Total', '@height'),
                                                            ('Data', '@source')])
show(ded_bar)
```
